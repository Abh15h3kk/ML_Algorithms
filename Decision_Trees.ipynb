# Train - Test

```python
X = pd.get_dummies(data=df.drop('species',axis=1),drop_first=True)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_predict = model.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
```

```python
confusion_matrix(y_test,base_predict)
```

```python
plot_confusion_matrix(model,X_test,y_test)
```

```python
print(classification_report(y_test,base_predict))
```

# Model result functions

```python
def report_model(model):
    model_predict = model.predict(X_test)
    print(classification_report(y_test, model_predict))
    print('\n')
    plt.figure(figsize=(12,8),dpi=200)
    plot_tree(model,filled=True,feature_names = X.columns)
```

```python
help(DecisionTreeClassifier)
```

```python
pruned = DecisionTreeClassifier(max_depth=2)
pruned.fit(X_train,y_train)
```

```python
report_model(pruned)
```

```python
pruned_tree = DecisionTreeClassifier(max_leaf_nodes=3)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train,y_train)
```

```python
report_model(entropy_tree)
```
